# Contour Detection

## External vs Internal Contours

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/internal_external.png', 0)
```

```python
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
image = img
```

```python
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
type(contours)
```

```python
len(contours)
```

```python
type(hierarchy)
```

```python
hierarchy
```

```python
external_contours = np.zeros(image.shape)
```

```python
external_contours.shape
```

```python
plt.imshow(image)
```

```python
for i in range(len(contours)):
    
    # External
    if hierarchy[0][i][3] == -1:
        
        cv2.drawContours(external_contours, contours, i, 255, -1)
```

```python
plt.imshow(external_contours, cmap='gray')
```

```python
internal_contours = np.zeros(image.shape)

for i in range(len(contours)):
    
    # External
    if hierarchy[0][i][3] != -1:
        
        cv2.drawContours(internal_contours, contours, i, 255, -1)
```

```python
plt.imshow(internal_contours, cmap='gray')
```

```python
ham_contours = np.zeros(image.shape)

for i in range(len(contours)):
    
    # External
    if hierarchy[0][i][3] == 4:
        
        cv2.drawContours(ham_contours, contours, i, 255, -1)
```

```python
plt.imshow(ham_contours, cmap='gray')
```
